## **SQL statement untuk menemukan jumlah customer per country**

```python
sql6 = """SELECT customer_state, COUNT(customer_id) as total_customer
          FROM public.oilist_customer
          GROUP BY customer_state
          ORDER BY total_customer DESC;"""
          ## JIKA DARI DATA TERKECIL GUNAKAN ASC, ATAU DATA RANDOM/TDK BERURUT : HAPUS ARGUMEN ORDER BY ##
jml_cust_per_country = pd.read_sql_query(sql6, conn)
jml_cust_per_country
```

## **SQL statement untuk menemukan jumlah total order per country**

```python
sql7 = """SELECT data1.customer_state, COUNT(data2.order_id) as jumlah_order
          FROM public.oilist_customer data1
          INNER JOIN public.oilist_orders data2
          ON data1.customer_id = data2.customer_id
          GROUP BY data1.customer_state
          ORDER BY jumlah_order ASC;"""
          #State dgn jumlah order terendah adalah "RR" dan tertinggi "SP"
total_order_per_country = pd.read_sql_query(sql7,conn)
total_order_per_country
```

## **SQL statement untuk menemukan berapa banyak jumlah order yang diterima oleh seller dalam interval 1 bulan**

```python
sql8 = """SELECT data2.seller_id, CAST(EXTRACT(MONTH FROM TO_DATE( data1.order_purchase_timestamp, 'DD/MM/YYYY')) AS INT) as order_month , COUNT(data1.order_id) as jumlah_order
          FROM public.oilist_orders data1
          INNER JOIN public.oilist_orders_detail data2
          ON data1.order_id = data2.order_id
          GROUP BY data2.seller_id, data1.order_purchase_timestamp, data1.order_id
          ORDER BY data2.seller_id, order_month DESC ;"""
jml_order_1_bulan = pd.read_sql_query(sql8,conn)
jml_order_1_bulan
```

## **SQL statement untuk menemukan TOP 10 buyer by total_price dari setiap country**

```python
sql9 = """SELECT data2.customer_id, data1.customer_state, SUM(data3.price) as total_price
          FROM public.oilist_customer data1
            INNER JOIN public.oilist_orders data2
            ON data1.customer_id = data2.customer_id
            INNER JOIN public.oilist_orders_detail data3
            ON data2.order_id = data3.order_id
          GROUP BY data2.customer_id, data1.customer_state, data3.price
          ORDER BY total_price DESC
          LIMIT 10
           ;"""
total_price = pd.read_sql_query(sql9,conn)
total_price
```

## **SQL statement untuk menemukan latest order_purchase_timestamp dari setiap buyer**

```python
sql10 = """ SELECT data.customer_id, MAX(TO_TIMESTAMP(order_purchase_timestamp, 'DD/MM/YYYY HH24.MI')) as order_purchase_timestamp
            FROM public.oilist_orders data
            GROUP BY data.customer_id, order_purchase_timestamp
            ORDER BY order_purchase_timestamp DESC
            ;"""
order_purchases_timestamp = pd.read_sql_query(sql10,conn)
order_purchases_timestamp
```

## **SQL statement untuk menemukan jumlah masing-masing order_status dari masing-masing customer**

```python
sql11 = """SELECT customer_id, order_status, COUNT(order_id)
           FROM public.oilist_orders 
           GROUP BY customer_id, order_status
           ORDER BY customer_id ;"""
order_status = pd.read_sql_query(sql11,conn)
order_status
```
